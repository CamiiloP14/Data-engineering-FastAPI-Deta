Vamos a realizar un análisis exploratorio de datos, primero realizamos una ingesta de los datos desde las diversas fuentes, para posteriormente aplicar las transformaciones solicitadas.

```python
#importamos librerias necesarias

import pandas as pd
```

```python
#leemos los datasets

amzn = pd.read_csv(r'Datasets/amazon_prime_titles-score.csv')
disney =pd.read_csv(r'Datasets/disney_plus_titles-score.csv')
hulu =pd.read_csv(r'Datasets/hulu_titles-score (2).csv')
netflix =pd.read_csv(r'Datasets/netflix_titles-score.csv')
```

Hacemos las transformaciones necesarias.

Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)



```python
amzn['id'] = 'a' +amzn['show_id']
disney['id'] = 'd' +disney['show_id']
hulu['id'] = 'h' +hulu['show_id']
netflix['id'] = 'n' +netflix['show_id']
```

Los valores nulos del campo rating se van a reemplazar por el string “G” (corresponde al maturity rating: “general for all audiences”

```python
amzn['rating'].fillna('g',inplace=True)
disney['rating'].fillna('g',inplace=True)
hulu['rating'].fillna('g',inplace=True)
netflix['rating'].fillna('g',inplace=True)
```

De haber fechas, deberán tener el formato AAAA-mm-dd

```python
amzn.head()
```

La columna 'date_added' es fecha. Vamos a cambiarla a tipo datetime con el formato 'AAAA-mm-dd'.

```python
amzn['date_added'] = pd.to_datetime(amzn['date_added'], infer_datetime_format=True)
disney['date_added'] = pd.to_datetime(disney['date_added'], infer_datetime_format=True)
hulu['date_added'] = pd.to_datetime(hulu['date_added'], infer_datetime_format=True)
netflix['date_added'] = pd.to_datetime(netflix['date_added'], infer_datetime_format=True)
```

El campo duration lo vamos a separar en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas).

Primero eliminamos los espacios finales de cada lado de la columna.

```python
amzn['duration']=amzn['duration'].str.lstrip()
amzn['duration']=amzn['duration'].str.rstrip()

disney['duration']=disney['duration'].str.lstrip()
disney['duration']=disney['duration'].str.lstrip()

hulu['duration']=hulu['duration'].str.lstrip()
hulu['duration']=hulu['duration'].str.lstrip()

netflix['duration']=netflix['duration'].str.lstrip()
netflix['duration']=netflix['duration'].str.lstrip()
```

```python
sep_amzn=amzn['duration'].str.split(" ", n = 1, expand = True) 
sep_hulu=hulu['duration'].str.split(" ", n = 1, expand = True)
sep_disney=disney['duration'].str.split(" ", n = 1, expand = True)
sep_netflix=netflix['duration'].str.split(" ", n = 1, expand = True)
```

```python
amzn['duration_int']=sep_amzn[0]
amzn['duration_type']=sep_amzn[1]

disney['duration_int']=sep_disney[0]
disney['duration_type']=sep_disney[1]

hulu['duration_int']=sep_hulu[0]
hulu['duration_type']=sep_hulu[1]

netflix['duration_int']=sep_netflix[0]
netflix['duration_type']=sep_netflix[1]
```

En la columna 'duration_type' cambiamos la palabra 'Seasons' por 'Season'

```python
amzn['duration_type'].unique()
```

```python
amzn['duration_type'].replace({'Seasons':'Season'}, inplace=True)
disney['duration_type'].replace({'Seasons':'Season'}, inplace=True)
hulu['duration_type'].replace({'Seasons':'Season'}, inplace=True)
netflix['duration_type'].replace({'Seasons':'Season'}, inplace=True)
```

```python
amzn['duration_type'].unique()
```

cambios el tipo de la columna 'duration_type' a tipo int.

```python
amzn['duration_int'] = pd.to_numeric(amzn['duration_int'], errors='coerce')
disney['duration_int'] = pd.to_numeric(disney['duration_int'], errors='coerce')
hulu['duration_int'] =hulu['duration_int'].fillna(0)
hulu['duration_int'] =pd.to_numeric(hulu['duration_int'], errors='coerce')
netflix['duration_int'] = pd.to_numeric(netflix['duration_int'], errors='coerce')
```

```python
concat_1 = pd.concat([amzn, disney], axis=0)
concat_2=pd.concat([hulu,netflix], axis=0)
movies=pd.concat([concat_1,concat_2], axis=0)
```

```python
movies.info()
```

Ahora normalizamos los campos de texto con minúsculas, sin excepciones

```python
movies.head()
```

```python
movies['show_id']=movies['show_id'].str.lower()
movies['type']=movies['type'].str.lower()
movies['title']=movies['title'].str.lower()
movies['director']=movies['director'].str.lower()
movies['cast']=movies['cast'].str.lower()
movies['country']=movies['country'].str.lower()
movies['rating']=movies['rating'].str.lower()
movies['duration']=movies['duration'].str.lower()
movies['listed_in']=movies['listed_in'].str.lower()
movies['description']=movies['description'].str.lower()
movies['duration_type']=movies['duration_type'].str.lower()
```

```python
movies.info()
```

Finalmente exportamos el archivo en formato csv, el cual nos serivirá para hacer las consultas en nuesta API.

```python
movies.to_csv('movies.csv', index=False)
```
